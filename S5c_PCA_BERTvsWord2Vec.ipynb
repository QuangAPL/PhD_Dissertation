##### Comparison Between BERT and Word2Vec Embeddings

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Load BERT embeddings
bert_embeddings = pd.read_csv('vectorized_bert_300.csv')

# Load Word2Vec embeddings
word2vec_embeddings = pd.read_csv('vectorized_word2vec_300.csv')

# Initialize scaler
scaler = StandardScaler()

# Normalize both embeddings
bert_embeddings_scaled = scaler.fit_transform(bert_embeddings)
word2vec_embeddings_scaled = scaler.fit_transform(word2vec_embeddings)

# Apply PCA
bert_pca = PCA(n_components=2).fit_transform(bert_embeddings_scaled)
word2vec_pca = PCA(n_components=2).fit_transform(word2vec_embeddings_scaled)

# Convert to DataFrame
bert_pca_df = pd.DataFrame(bert_pca, columns=['PC1', 'PC2'])
word2vec_pca_df = pd.DataFrame(word2vec_pca, columns=['PC1', 'PC2'])

# Plot the results
fig, axs = plt.subplots(1, 2, figsize=(20, 7))

sns.scatterplot(x='PC1', y='PC2', data=bert_pca_df, ax=axs[0])
axs[0].set_title('PCA of BERT Embeddings')

sns.scatterplot(x='PC1', y='PC2', data=word2vec_pca_df, ax=axs[1])
axs[1].set_title('PCA of Word2Vec Embeddings')

plt.show()
```
